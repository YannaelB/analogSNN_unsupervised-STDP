# Unsupervised Learning - STDP
## About this code:

In this code, we implement a Spiking Neural Network for solving every logical operators with eNeuron ML. The used method is the exact same as 'XOR_rate_code_unsupervised_STDP_lateralWTA.ipynb' which is a visual representation of the code 'hyperparameter_architecture/unsupervised_classic_STDP.py'.

This code utilises the architecture : [2,13]+exci+inhi+wta. It means that there is two input neurons fully connected
with 13 outputs neurons with both excitatory and inhibitory synapses conducted with STDP rules. 
Plus, there is a lateral Winner-Take-All synaptic connection between output neurons. It implies 208 synapses.
It utilises ''neuron_eqs_v1''. However, rather than stopping the training once the XOR is solved, we stop the training once the output layer achieve to differenciate each input.

Results : The network achieves 100% accuracy with a widely dispersed number of epochs. It has some difficulties with the input [0,0]. Indeed it would require that the input [0,0] makes at least one neurons spiking much more than for the other inputs (which is feasible) + it requires that this neuron fires at a higher rate than the other neuron for the input [0,0] which is more complitcated according to 'XOR_rate_code_unsupervised_STDP_lateralWTA.ipynb'. However this is not impossible.

# Model

```python
from brian2 import clear_cache
from brian2 import device
device.reinit()
start_scope()


# Simulation parameters
defaultclock.dt = 0.005*us  



v_rest = -80*mV
v_threshold = 60*mV  
v_std_thresh = 0*mV #standart deviation for the threshold value (randn :drawn from the standard normal distribution, i.e. with mean 0 and standard deviation 1) 
v_reset = -100*mV
v0 = -40*mV  
tau = 0.5*ms
tau2 = 15*us
Rm = 5000*Mohm
I0=1*amp
a_v = 1126.6343973469338
b_v = 169200.45773494235
x_threshold_v= 8.3e-11*amp
avoid_error = 0.01*pA
a_value = 6.23667974e13
refrac = 0*us


n_input, n_output = 2,13

tau_stdp = 1.88116837
mag_stdp = 0.007070372048
time_sample = 115
time_resting = 20
alpha_decay = 1
alpha_e = 0.15
alpha_i = 0.04
beta_e = 1.9
beta_i = 0.5
avoid_error = 0.01*pA
c = 1

neuron_eqs_v1 = '''
dv/dt =  int((I + Ie+I_inhi) > x_threshold_v) *(v_rest - v) / tau2 + (( int((I + Ie+I_inhi) > x_threshold_v) *((a_v + b_v * log((1-int((I + Ie+I_inhi) > x_threshold_v))+int((I + Ie+I_inhi) > x_threshold_v)*(I+Ie+I_inhi+avoid_error)/(x_threshold_v) ))*I0)/a_value)*Rm) / tau : volt (unless refractory)
Ie =  activity*alpha_e*ge * uS *(80*mV -c*v)*inhi_factor           : amp
I_inhi =  activity*alpha_i*gi * uS *(-120*mV -c*v)           : amp
I : amp
dge/dt = -ge/(beta_e*us)            : 1
dgi/dt = -gi/(beta_i*us)            : 1
inhi_factor : 1
activity :1
'''


V_a, V_b= 70*mV,90*mV
R_a = Rm*(V_a/mV)/(5/6)/(abs(v_threshold/mV)+abs(v_reset/mV))  #New resistance for rising till V_a
R_b = Rm*(V_b/mV)/(1/6)/(abs(v_threshold/mV)+abs(v_reset/mV)) #New resistance for rising till V_b 
neuron_eqs_v3 = '''
dv/dt =  int((I + Ie+I_inhi) > x_threshold_v) *(v_rest - v) / tau2 + (( int((I + Ie+I_inhi) > x_threshold_v) *((a_v + b_v * log((1-int((I + Ie+I_inhi) > x_threshold_v))+int((I + Ie+I_inhi) > x_threshold_v)*(I+Ie+I_inhi+avoid_error)/(x_threshold_v) ))*I0)/a_value)*Rm) / tau : volt (unless refractory)
Ie =  activity*alpha_e*ge * uS *(80*mV -c*v)*inhi_factor           : amp
I_inhi =  activity*alpha_i*gi * uS *(-120*mV -c*v)           : amp
I : amp
dge/dt = -ge/(beta_e*us)            : 1
dgi/dt = -gi/(beta_i*us)            : 1
inhi_factor : 1
Rm = (R_a)*int(v<(v_reset+V_a)) + (R_b)*int(v>=(v_reset+V_a)) : ohm
activity :1
'''

reset_eqs='''
v = v_reset
I=0*nA
'''


#STDP parameters
taupre = tau_stdp*us
taupost = taupre
wmax = 1
dApre = mag_stdp
dApost = -dApre * taupre / taupost * 1.05

stdp_eqs_exci = '''
    w : 1
    alpha : 1 (shared)
    lr : 1 (shared)
    dApre/dt = -Apre / taupre : 1 (event-driven)
    dApost/dt = -Apost / taupost : 1 (event-driven)
'''
pre_eqs_exci='''
    ge_post += w
    Apre += dApre
    w = clip(alpha*w + lr*Apost, 0, wmax)
'''
post_eqs_exci='''
    Apost += dApost
    w = clip(w + lr*Apre, 0, wmax)
'''

stdp_eqs_inhi = '''
    w : 1
    alpha : 1 (shared)
    lr : 1 (shared)
    dApre/dt = -Apre / taupre : 1 (event-driven)
    dApost/dt = -Apost / taupost : 1 (event-driven)
'''
pre_eqs_inhi='''
    gi_post += w
    Apre += dApre
    w = clip(alpha*w + lr*Apost, 0, wmax)
'''
post_eqs_inhi='''
    Apost += dApost
    w = clip(w + lr*Apre, 0, wmax)
'''

synapse_model = "w : 1"

```

# Network

```python
from brian2 import clear_cache
from brian2 import device
device.reinit()
start_scope()



input_group = NeuronGroup(n_input, neuron_eqs_v1, threshold='v >= v_threshold + v_std_thresh*randn()', reset='v = v_reset', refractory=0*us, method='heun')
input_group.inhi_factor = 1
input_group.activity = 1


output_group = NeuronGroup(n_output, neuron_eqs_v1, threshold='v >= v_threshold + v_std_thresh*randn()', reset='v = v_reset', refractory=0*us, method='heun')
output_group.inhi_factor = 1
output_group.activity = 1



input_sy_exci = Synapses(input_group, output_group, model=stdp_eqs_exci, on_pre=pre_eqs_exci, on_post=post_eqs_exci)
input_sy_exci.connect(True)
input_sy_exci.lr = 0.5
input_sy_exci.w = 'rand() * wmax * 0.8 + 0.1*wmax'
input_sy_exci.delay = 'rand()*0*us'

input_sy_inhi = Synapses(input_group, output_group, model=stdp_eqs_inhi, on_pre=pre_eqs_inhi, on_post=post_eqs_inhi)
input_sy_inhi.connect(True)
input_sy_inhi.lr = 0.5
input_sy_inhi.w = 'rand() * wmax * 0.5 + 0.1*wmax'
input_sy_inhi.delay = 'rand()*0*us'

wta_sy = Synapses(output_group, output_group, model=synapse_model, on_pre='gi_post += w')
wta_sy.connect(condition='i != j')
wta_sy.w = 'rand() * wmax * 1.2 + 0.8*wmax'
wta_sy.delay = 'rand()*0*us'

input_sy_exci.alpha = alpha_decay
input_sy_inhi.alpha = alpha_decay



# Monitors for input neurons
input_spike_monitor = SpikeMonitor(input_group)

# Monitors for output neuron
output_spike_monitor = SpikeMonitor(output_group)


print("done")

time_per_sample = time_sample * us
resting_time = time_resting * us
input_group.v = v_reset
output_group.v = v_reset

#Create the network
net = Network(input_group, output_group, 
              input_sy_exci, input_sy_inhi,wta_sy,
              input_spike_monitor,output_spike_monitor,
             )

```

## Creating XOR Data 


```python

def batch_data_XOR(n_train,n_test):
    X_train = n_train*[[0,0]] + n_train*[[0,1]] + n_train*[[1,0]] + n_train*[[1,1]]
    shuffle(X_train)
    y_train = [a ^ b for a, b in X_train]

    X_test = n_test*[[0,0]] + n_test*[[0,1]] + n_test*[[1,0]] + n_test*[[1,1]]
    shuffle(X_test)
    y_test = [a ^ b for a, b in X_test]
    
    return X_train,y_train,X_test,y_test

def batch_data_NOR(n_train,n_test):
    X_train = n_train*[[0,0]] + n_train*[[0,1]] + n_train*[[1,0]] + n_train*[[1,1]]
    shuffle(X_train)
    y_train = [int(not (a or b)) for a, b in X_train]

    X_test = n_test*[[0,0]] + n_test*[[0,1]] + n_test*[[1,0]] + n_test*[[1,1]]
    shuffle(X_test)
    y_test = [int(not (a or b)) for a, b in X_test]
    
    #print("NOR : X_test = ", X_test, " y_test = ", y_test)
    return X_train,y_train,X_test,y_test

def batch_data_NAND(n_train,n_test):
    X_train = n_train*[[0,0]] + n_train*[[0,1]] + n_train*[[1,0]] + n_train*[[1,1]]
    shuffle(X_train)
    y_train = [int(not (a and b)) for a, b in X_train]

    X_test = n_test*[[0,0]] + n_test*[[0,1]] + n_test*[[1,0]] + n_test*[[1,1]]
    shuffle(X_test)
    y_test = [int(not (a and b)) for a, b in X_test]
    
    #print("NAND : X_test = ", X_test, " y_test = ", y_test)
    return X_train,y_train,X_test,y_test

batch_data_NOR(2,2)

def from_bit_to_current(x):
    if x == [0,0]:
        return [0.3,0.3]
    elif x == [0,1]:
        return [0.3,2]
    elif x == [1,0]:
        return [2,0.3]
    elif x == [1,1]:
        return [2,2]
    else:
        print(" WARNING PROBLEM !")
        
def from_bit_to_label_2(x):
    if x == [0,0]:
        return 0
    elif x == [0,1]:
        return 1
    elif x == [1,0]:
        return 1
    elif x == [1,1]:
        return 0
    else:
        print(" WARNING PROBLEM !")
        
def from_bit_to_label_4(x):
    if x == [0,0]:
        return 0
    elif x == [0,1]:
        return 1
    elif x == [1,0]:
        return 2
    elif x == [1,1]:
        return 3
    else:
        print(" WARNING PROBLEM !")
        
def from_label_to_XOR(x):
    if x == 0:
        return 0
    elif x == 1:
        return 1
    elif x == 2:
        return 1
    elif x == 3:
        return 0
    else:
        print(" WARNING PROBLEM !")
        

def from_label_to_NAND(x):
    if x == 0:
        return 1
    elif x == 1:
        return 1
    elif x == 2:
        return 1
    elif x == 3:
        return 0
    else:
        print(" WARNING PROBLEM !")
        
def from_label_to_NOR(x):
    if x == 0:
        return 1
    elif x == 1:
        return 0
    elif x == 2:
        return 0
    elif x == 3:
        return 0
    else:
        print(" WARNING PROBLEM !")
```

# Training Part


```python


def training(idx_epoch,X_train,y_train,plot=False):
    
    input_sy_exci.lr = 1
    input_sy_inhi.lr = 1

    spikes4 = np.zeros((4, n_output))
    

    old_spike_counts = np.zeros(n_output)
    old_spike_counts = np.copy(output_spike_monitor.count)
    
    for j, (sample, label) in enumerate(zip(X_train, y_train)):
        
        input_group.I = from_bit_to_current(sample) * nA
        input_group.v = v_rest
        output_group.v = v_rest
        # Start the training simulation
        net.run(time_per_sample)

        counter_spike = output_spike_monitor.count - old_spike_counts
        spikes4[from_bit_to_label_4(sample)] += counter_spike 
               
          
        # Allow the variables to return to their resting values
        input_group.I = 0 * nA
        net.run(resting_time) 
        input_group.v = v_rest
        output_group.v = v_rest
        old_spike_counts = np.copy(output_spike_monitor.count)
        
        if j >= 8 and j<9 and idx_epoch<1 and plot:
            print("here in plot burst map")
            plt.figure(figsize=(20, 10)) 
            plot(input_spike_monitor.t/us, input_spike_monitor.i+np.array([-n_input for i in range(len(input_spike_monitor.i))]), '.b',label='input')
            plot(output_spike_monitor.t/us, output_spike_monitor.i, '.r',label='output')
            plt.ylabel('Neuron index')
            plt.legend()
            plt.show()
            
        #NORMALISATION :
        weight_matrix = np.zeros([n_input, n_output])
        weight_matrix[input_sy_exci.i, input_sy_exci.j] = input_sy_exci.w
        sumCol = np.sum(weight_matrix, axis=0)
        colFactors = 1/sumCol
        for jj in range(n_output):
            weight_matrix[:,jj] *= colFactors[jj]
        input_sy_exci.w = weight_matrix[input_sy_exci.i, input_sy_exci.j]

        weight_matrix = np.zeros([n_input, n_output])
        weight_matrix[input_sy_inhi.i, input_sy_inhi.j] = input_sy_inhi.w
        sumCol = np.sum(weight_matrix, axis=0)
        colFactors = 0.5/sumCol
        for jj in range(n_output):
            weight_matrix[:,jj] *= colFactors[jj]
        input_sy_inhi.w = weight_matrix[input_sy_inhi.i, input_sy_inhi.j]


        return print("training done")
```

# Labelisation

```python
def labelisation(idx_epoch,X_train,y_train):
    input_sy_exci.lr = 0
    input_sy_inhi.lr = 0

    spikes4 = np.zeros((4, n_output))

    old_spike_counts = np.zeros(n_output)
    old_spike_counts = np.copy(output_spike_monitor.count)


    label_percent = 0.5
    nb_label = int(label_percent*len(y_train))
    X_labeling = X_train[:nb_label]
    y_labeling = y_train[:nb_label]


    for j, (sample, label) in enumerate(zip(X_labeling, y_labeling)):
    
        input_group.I = from_bit_to_current(sample) * nA
        input_group.v = v_rest
        output_group.v = v_rest

        # Start the simulation
        net.run(time_per_sample)

        counter_spike = output_spike_monitor.count - old_spike_counts
        spikes4[from_bit_to_label_4(sample)] += counter_spike    

        # Allow the variables to return to their resting values
        input_group.I = 0 * nA
        net.run(resting_time) 
        input_group.v = v_rest
        output_group.v = v_rest

        old_spike_counts = np.copy(output_spike_monitor.count)
                                        
    return spikes4


    
```

```python
def plot_post_label(spikes4,plot=False):
    if plot:
        plt.figure()
        label_list = [[0,0],[0,1],[1,0],[1,1]]
        for i in range(len(spikes4)):
            if spikes4[i].any() != np.zeros(len(spikes4[i])).any():
                plt.plot(range(len(spikes4[i])),spikes4[i], label=f'label {label_list[i]}')
        plt.legend()
        plt.show()
    
     
    labeled_neurons4 = np.argmax(spikes4, axis=0)
    redundancy_neuron4 = np.bincount(labeled_neurons4, minlength=5)
    if plot:
        plt.figure()
        bar(range(len(redundancy_neuron4)),redundancy_neuron4)
        xticks(range(len(redundancy_neuron4)))
        print(" labeled_neurons = ", labeled_neurons4)
        print(" redundancy_neuron = ", redundancy_neuron4)
    
    return labeled_neurons4,redundancy_neuron4
    
    
```

# Evaluation

```python
def evaluationXOR(labeled_neurons4,redundancy_neuron4,idx_epoch,X_test,y_test,plot=False):
    input_sy_exci.lr = 0
    input_sy_inhi.lr = 0

    spikes4 = np.zeros((4, n_output))

    old_spike_counts = np.zeros(n_output)
    old_spike_counts = np.copy(output_spike_monitor.count)
    
    num_correct_output_max4 = 0

    for j, (sample, label) in enumerate(zip(X_test, y_test)):


        input_group.v = v_rest
        output_group.v = v_rest
        input_group.I = from_bit_to_current(sample) * nA
        net.run(time_per_sample)
    
        counter_spike = output_spike_monitor.count - old_spike_counts
        spikes4[from_bit_to_label_4(sample)] += counter_spike
        

        output_label_max4 = labeled_neurons4[np.argmax(counter_spike)]
        
    
        if j < 3:
            pass
            print("output_label_max4 = ", output_label_max4, " and int(label) = ", int(label))
            
        if output_label_max4 == from_bit_to_label_4(sample):
            num_correct_output_max4 += 1
       
        
        input_group.I = 0 * nA
        net.run(resting_time)
        input_group.v = v_rest
        output_group.v = v_rest
        old_spike_counts = np.copy(output_spike_monitor.count)
    
   
    accuracy = (num_correct_output_max4 / len(X_test))
   

    return accuracy

```

# 3D XOR solvation 

0 binary <=> 0.3*nA => frequency = ..kHz 
input 0.5 <=> (0.3+1.5*0.5)=1.15*nA => frequency = ..kHz
1 binary <=> 2.0*nA => frequency = ..kHz 

Par simplification, je raisonne en courant mais il faudrait raisonner en fréquence pour graph


```python
def input_3D_vec(n_point = 5):
    
    binary_list = [i / (n_point - 1) for i in range(n_point)]

    input_3D = np.zeros((2,len(binary_list)**2))
    X_3D,Y_3D = [],[]
    input_3D_current = []

    k = 0
    for i,binary_x in enumerate(binary_list):
        for j,binary_y in enumerate(binary_list):
            input_3D[0,k], input_3D[1,k] = binary_x,binary_y
            X_3D.append(binary_x)
            Y_3D.append(binary_y)
            input_3D_current.append([0.3+1.7*binary_x,0.3+1.7*binary_y])
            k+=1

    return X_3D,Y_3D,input_3D_current

def plot_3D(labeled_neurons4,idx_epoch,n_point = 5,plot=False):
    
    input_sy_exci.lr = 0
    input_sy_inhi.lr = 0
    
    X_3D,Y_3D,input_3D_current = input_3D_vec(n_point)
    
    Z_3D_XOR,Z_3D_NAND,Z_3D_NOR = [],[],[]

    old_spike_counts = np.zeros(n_output)
    old_spike_counts = np.copy(output_spike_monitor.count)


    for j, (sample) in enumerate(input_3D_current):

        input_group.v = v_rest
        output_group.v = v_rest
        input_group.I = sample * nA
        net.run(time_per_sample)
    
        counter_spike = output_spike_monitor.count - old_spike_counts
        
        output_label_max4_XOR = from_label_to_XOR(labeled_neurons4[np.argmax(counter_spike)])
        output_label_max4_NAND = from_label_to_NAND(labeled_neurons4[np.argmax(counter_spike)])
        output_label_max4_NOR = from_label_to_NOR(labeled_neurons4[np.argmax(counter_spike)])
        Z_3D_XOR.append(output_label_max4_XOR)
        Z_3D_NAND.append(output_label_max4_NAND)
        Z_3D_NOR.append(output_label_max4_NOR)
        
        
        input_group.I = 0 * nA
        net.run(resting_time) 
        input_group.v = v_rest
        output_group.v = v_rest
        old_spike_counts = np.copy(output_spike_monitor.count)

     
    print("len Z_3D_XOR = ",len(Z_3D_XOR))
    ### PLOT 3D graph:

    # Given vectors
    x = np.array(X_3D)
    y = np.array(Y_3D)
    z_xor = np.array(Z_3D_XOR)
    z_nand = np.array(Z_3D_NAND)
    z_nor = np.array(Z_3D_NOR)

    # Create a grid for the surface, this time ensuring x goes from 1 to 0
    xi = np.linspace(x.max(), x.min(), 100)
    yi = np.linspace(y.min(), y.max(), 100)
    Xi, Yi = np.meshgrid(xi, yi)

    # Linearly interpolate the z values on the grid
    Zi_xor = griddata((x, y), z_xor, (Xi, Yi), method='linear')
    Zi_nand = griddata((x, y), z_nand, (Xi, Yi), method='linear')
    Zi_nor = griddata((x, y), z_nor, (Xi, Yi), method='linear')

    # Plotting the surface graph
    fig_surf = plt.figure()
    ax_surf = fig_surf.add_subplot(111, projection='3d')
    ax_surf.plot_surface(Xi, Yi, Zi_xor, cmap='viridis', edgecolor='none')
    ax_surf.set_xlim(1, 0)
    ax_surf.set_title('XOR 3D Surface Plot')
    ax_surf.set_xlabel('X')
    ax_surf.set_ylabel('Y')
    ax_surf.set_zlabel('Z')
    plt.show()
    
    # Plotting the surface graph
    fig_surf = plt.figure()
    ax_surf = fig_surf.add_subplot(111, projection='3d')
    ax_surf.plot_surface(Xi, Yi, Zi_nand, cmap='viridis', edgecolor='none')
    ax_surf.set_title('NOR 3D Surface Plot')
    ax_surf.set_xlabel('X')
    ax_surf.set_ylabel('Y')
    ax_surf.set_zlabel('Z')
    plt.show()
    
    # Plotting the surface graph
    fig_surf = plt.figure()
    ax_surf = fig_surf.add_subplot(111, projection='3d')
    ax_surf.plot_surface(Xi, Yi, Zi_nor, cmap='viridis', edgecolor='none')
    ax_surf.set_xlim(1, 0)
    ax_surf.set_title('XOR 3D Surface Plot')
    ax_surf.set_xlabel('X')
    ax_surf.set_ylabel('Y')
    ax_surf.set_zlabel('Z')
    plt.show()

```

# Run Network training

```python

epoch = 15

weight_evol_exci = []
weight_evol_inhi = []
accuracy_evol = []
n_train,n_test = 10,3

nb_epoch_accuracy100 = []
for _ in range(1):
    seed()
    tic = time.time()
    for idx_epoch in tqdm(range(epoch)):
        X_train,y_train,X_test,y_test = batch_data_XOR(n_train,n_test)
        training(idx_epoch,X_train,y_train,plot=False)
        spikes4 = labelisation(idx_epoch,X_train,y_train)
        labeled_neurons4,redundancy_neuron4 = plot_post_label(spikes4,plot=False)
        accuracy = evaluationXOR(labeled_neurons4,redundancy_neuron4,idx_epoch,X_test,y_test,plot=False)
        accuracy_evol.append(accuracy)
        if accuracy >= 1:
            break
    print("It has taken = ", idx_epoch+1, "epochs for accuracy:", accuracy*100,"% in ",time.time()-tic,"s \n\n")
    nb_epoch_accuracy100.append(idx_epoch+1)
    
    
print(" nb_epoch_accuracy100 = ",nb_epoch_accuracy100)
print(np.mean(nb_epoch_accuracy100))
```

## Accuracy evol

```python

plt.figure(figsize=(10, 6)) 

plt.plot(range(len(accuracy_evol)),accuracy_evol)

plt.title(' accuracy evolution ')
plt.xlabel(f' epoch  ')
plt.ylabel(' accuracy ')
plt.grid(True)
plt.show()
```

# NAND

```python
def evaluationNAND(labeled_neurons4,redundancy_neuron4,idx_epoch,X_test,y_test,plot=False):
    input_sy_exci.lr = 0
    input_sy_inhi.lr = 0

    spikes4 = np.zeros((4, n_output))

    old_spike_counts = np.zeros(n_output)
    old_spike_counts = np.copy(output_spike_monitor.count)
    
    num_correct_output_max4 = 0

    for j, (sample, label) in enumerate(zip(X_test, y_test)):


        input_group.v = v_rest
        output_group.v = v_rest
        input_group.I = from_bit_to_current(sample) * nA
        net.run(time_per_sample)
    
        counter_spike = output_spike_monitor.count - old_spike_counts
        spikes4[from_bit_to_label_4(sample)] += counter_spike
        print("labeled_neurons4 = ", labeled_neurons4)
        print(" for sample = ",sample,"counter_spike = ",counter_spike,"\n")

        output_label_max4 = from_label_to_NAND(labeled_neurons4[np.argmax(counter_spike)])
        
    
        if j < 10:
            pass
            print("output_label_max4 = ", output_label_max4, " and int(label) = ", int(label))
            
        if output_label_max4 == int(label):
            num_correct_output_max4 += 1
       
        input_group.I = 0 * nA
        net.run(resting_time)
        input_group.v = v_rest
        output_group.v = v_rest
        old_spike_counts = np.copy(output_spike_monitor.count)
    
    accuracy = (num_correct_output_max4 / len(X_test))
    return accuracy


epoch = 1

weight_evol_exci = []
weight_evol_inhi = []
accuracy_evol = []
n_train,n_test = 10,2

for _ in range(1):
    seed()
    tic = time.time()
    for idx_epoch in tqdm(range(epoch)):
        X_train,y_train,X_test,y_test = batch_data_NAND(n_train,n_test)
        accuracy = evaluationNAND(labeled_neurons4,redundancy_neuron4,idx_epoch,X_test,y_test,plot=False)
        accuracy_evol.append(accuracy)
        if accuracy >= 1:
            print("100% accuracy achieved but continue to 9 epoch")
            break
    print("It has taken = ", idx_epoch+1, "epochs for accuracy:", accuracy*100,"% in ",time.time()-tic,"s \n\n")
    
    #plot_3D(labeled_neurons4,idx_epoch,plot=True)
        

```
